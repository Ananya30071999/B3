1. Write a Python program to check if a given number is positive, negative, or zero using an `if-elif-else` statement.

2. Write a program that takes a list of numbers as input and prints the square of each number using a `for` loop.

```python
#My Answer to the above question
user_no_of_list_elements = int(input("Enter the number of elements in list:"))
user_elements = []
for i in range(1,user_no_of_list_elements+1):
    user_elements.append(int(input("Enter element :")))
for j in range(0,len(user_elements)):
    print(user_elements[j]**2)
```

#### Intermediate-Level Assignment
1. Write a Python program to implement a simple calculator. The program should:
   - Accept two numbers and an operator (`+`, `-`, `*`, `/`).
   - Perform the operation and display the result.

```python
#My Answer to the above question
def calculator():
    num_1 = int(input("Enter 1st Number: "))
    num_2 = int(input("Enter 2nd Number: "))
    operator = int(input("Which Operation you choose: 1 for Sum, 2 for Difference, 3 for Product, 4 for division"))
    if operator == 1:
        print(num_1+num_2)
    elif operator == 2:
        print(num_1-num_2)
    elif operator == 3:
        print(num_1*num_2)
    elif operator == 4:
        print(num_1/num_2)
    else:
        print("Wrong Operator")

calculator()
        
```

2. Write a function named `fibonacci(n)` that returns the first `n` numbers in the Fibonacci sequence.

```python
#My Answer to the above question
def fibonacci():
    list = [0,1]
    n = int(input("Enter the number of elements: "))
    if n == 1:
        list = [0]
        print(list)
    elif n == 2:
        list = [0,1]
        print(list)
    elif n>2:
        for i in range (1,n-1):
            list.append(list[i-1]+list[i])
        print(list)
    else:
        print("Invalid Input")

fibonacci()
            
```
